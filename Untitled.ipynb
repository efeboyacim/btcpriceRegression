```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('BTCData.csv')
```

```python
def check_data(df,head=5):
    print(20*"-" + "Information".center(20) + 20*"-")
    print(df.info())
    print(20*"-" + "Data Shape".center(20) + 20*"-")
    print(df.shape)
    print("\n" + 20*"-" + "The First 5 Data".center(20) + 20*"-")
    print(df.head())
    print("\n" + 20 * "-" + "The Last 5 Data".center(20) + 20 * "-")
    print(df.tail())
    print("\n" + 20 * "-" + "Missing Values".center(20) + 20 * "-")
    print(df.isnull().sum())
    print("\n" + 40 * "-" + "Describe the Data".center(40) + 40 * "-")
    print(df.describe().T)
    
check_data(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# 2017 yılına kadar olan tarihleri içermeyen verileri seçin
df_filtered = df[df['Date'].dt.year > 2016]

# Orijinal DataFrame'i güncelleyin
df = df_filtered

# Filtrelenmiş veriyi gösterin
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# 'Date' sütununu indeks olarak belirleyin
df.set_index('Date', inplace=True)

# Fiyatları içeren sütunları seçin
prices = ['Close']

# Seçilen sütunları kullanarak grafiği çizin
df[prices].plot(figsize=(10, 6))
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
df.head(1)
```

```python
temp_column = df['Volume'].copy()  # 'Volume' sütununu bir geçici sütuna kopyalayalım
df['Volume'] = df['Close']  # 'Volume' sütununu 'Close' sütunuyla değiştirelim
df['Close'] = temp_column  # 'Close' sütununu geçici sütunla değiştirelim
df.rename(columns={'Volume': 'Volume1'}, inplace=True)
df.rename(columns={'Close': 'Volume'}, inplace=True)
df.rename(columns={'Volume1': 'Close'}, inplace=True)
```

```python
df.head(1)
df.iloc[-2] = df.iloc[-1]
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
nan_cells = df.isna()

# Her bir hücredeki NaN değerlerini sayalım
nan_counts = nan_cells.sum()

# Toplam NaN değer sayısını alalım
total_nan_count = nan_counts.sum()

# Her bir sütundaki NaN değer sayılarını gösterelim
print("Her bir sütundaki NaN değer sayıları:")
print(nan_counts)

# Toplam NaN değer sayısını gösterelim
print("\nToplam NaN değer sayısı:", total_nan_count)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import XGBClassifier
# XGBoost Regressor'ı oluştur
regressor = XGBRegressor()

# Regressor'ı eğit
regressor.fit(X_train, y_train)
```

```python
# NaN değerleri kontrol et
print("Eğitim Seti NaN Değer Kontrolü:")
print(y_train.isna().sum())

print("\nTest Seti NaN Değer Kontrolü:")
print(y_test.isna().sum())
```

```python
from sklearn.metrics import r2_score

y_pred = regressor.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
y_train_pred = regressor.predict(X_train)
y_test_pred = regressor.predict(X_test)

# Eğitim ve test setlerindeki gerçek değerlerle tahmin edilen değerler arasındaki farkları hesaplayın
train_errors = y_train - y_train_pred
test_errors = y_test - y_test_pred

# Eğitim ve test setlerindeki hataları gösteren bir grafik oluşturun
plt.figure(figsize=(10, 6))
plt.scatter(np.arange(len(y_train)), train_errors, label='Train Errors', alpha=0.7)
plt.scatter(np.arange(len(y_test)), test_errors, label='Test Errors', alpha=0.7)
plt.axhline(0, color='red', linestyle='--', linewidth=2, label='Zero Error Line')
plt.title('Training and Test Errors')
plt.xlabel('Data Points')
plt.ylabel('Errors')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Eğitim ve test setlerindeki tahminleri alın
y_train_pred = regressor.predict(X_train)
y_test_pred = regressor.predict(X_test)

# Eğitim seti için scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_train, y=y_train_pred)
plt.title('Gerçek Değerler vs. Tahmin Edilen Değerler (Eğitim Seti)')
plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')
plt.show()

# Test seti için scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_test_pred)
plt.title('Gerçek Değerler vs. Tahmin Edilen Değerler (Test Seti)')
plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')
plt.show()
```
